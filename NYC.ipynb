```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/Users/altmshkhan/Downloads/Large_NYC_Visualization_Dataset.csv')

# Count listings per neighborhood
neighborhood_counts = df['Neighborhood'].value_counts()

# Create a bar chart
plt.figure(figsize=(12, 6))
neighborhood_counts.plot(kind='bar', color='skyblue')
plt.title('Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import numpy as np

df = pd.read_csv('/Users/altmshkhan/Downloads/Large_NYC_Visualization_Dataset.csv')

# Check that 'price' and 'square_footage' columns exist in your dataset
if 'House_Prices_Median' in df.columns and 'Unemployment_Rate' in df.columns:
    # Set the dimensions of the figure
    plt.figure(figsize=(14, 8))

    # Create a scatter plot
    sns.scatterplot(x='Unemployment_Rate', y='House_Prices_Median', data=df, alpha=0.6)

    # Fit a regression line
    sns.regplot(x='Unemployment_Rate', y='House_Prices_Median', data=df, scatter=False, color='red')

    # Set titles and labels
    plt.title('House_Prices_Median vs. Unemployment_Rate', fontsize=16)
    plt.xlabel('Unemployment_Rate', fontsize=14)
    plt.ylabel('House_Prices_Median', fontsize=14)
    plt.grid()

    # Show the plot
    plt.tight_layout()
    plt.show()
else:
    print("The dataset does not contain 'price' or 'square_footage' columns.")
```

```python
if 'Neighborhood' in df.columns and 'Population_Thousands' in df.columns:
    # Aggregate population by neighborhood
    neighborhood_population = df.groupby('Neighborhood')['Population_Thousands'].sum().reset_index()
    
    # Use the neighborhood as index and the population as values
    heatmap_data = neighborhood_population.set_index('Neighborhood')
    
    # Create a heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(data=heatmap_data.T, annot=True, cmap='YlGnBu', fmt=".0f")  # Transpose to have Neighborhood as y-axis
    
    # Set titles and labels
    plt.title('Population Heatmap by Neighborhood', fontsize=16)
    plt.xlabel('Neighborhood', fontsize=14)
    plt.ylabel('Population in Thousands', fontsize=14)

    # Show the plot
    plt.show()
else:
    print("The dataset does not contain 'Neighborhood' or 'Population_Thousands' columns.")
```
